# Tutorial script on how to use the anomaly model

First, let us import both the standard and utility libraries and files. *util.py* contains helpful functions while *models.py* holds our model architecture. Every other imports are standard python libraries.

```python
import os
import json
import numpy as np
import pandas as pd
import h5py
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import util
from models import rnn_model, fit_model
from matplotlib import pyplot as plt
import seaborn as sns
plt.rcParams['text.usetex'] = True
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
# plt.rcParams['pdf.fonttype'] = 42
sns.set(color_codes=True)
sns.set_context("paper")
%matplotlib inline
```

### Load the configuration file and the data.

For this default model, I am loading the data as one and split into training and testing. However, for others wanting to train, you load the training data and divide it into training and testing using the *train_test_split* used below. The parameters in the *config.json* can all be supplied or the default values are used. We are supplying experiment name for the *load_data* function to enable us identify which data file to load. This *load_data* only loads data already stored in the folder. We used *vectorize_data* to process and store the data.

```python
# load the params file
params = 'config.json'
if os.path.isfile(params):
    params = json.load(open(params, mode='r'))

batch_size = params['batch-size']  # Batch size for training.
epochs = params['epochs']  # Number of epochs to train for.    
# lets load the data for training and testing our default model
exp_name = 'normal' # for training the normal profile. check config.json for other options
X_data, Y_data = util.load_data(exp_name)
train_size = params['train-size']
X_train, X_test, Y_train, Y_test = train_test_split(X_data,Y_data,train_size=train_size,random_state=50)
```

### Train the model 

When we are done training. We plot the loss and accuracy to visualize how the model performed on the training and validation datasets. For the sake of the project, the training *training loss* and *validation loss* will be in one figure while the *training accuracy* and *validation accuracy* will be in one figure.

```python
# train the model here
# we can use sklearn joblib to dump and reload these scalers as its to be used for both training and testing
scaler_x = MinMaxScaler() 
X_scaled = scaler_x.fit_transform(X_train.reshape(-1,X_train.shape[-1]))
X_scaled = X_scaled.reshape(X_train.shape[0],X_train.shape[1],X_train.shape[2])
scaler_y = MinMaxScaler()
Y_scaled = scaler_y.fit_transform(Y_train)
```

```python
# fit the model
model, history = fit_model(X_scaled, Y_scaled)
```

### Visualize the training stats

This section is a format in which stats are plotted after training. I can also put the training in a loop so that we stream these stats every epoch. Again, for your project, I think the loop version will be better. We will discuss this further.

```python
# lets visualize the training and validation history
history_df = pd.DataFrame.from_dict(history.history)
fig, axes = plt.subplots(2,1,figsize=(10,10),dpi=300,sharex='col')
loss = history_df['loss']
val_loss = history_df['val_loss']
acc = history_df['accuracy']
val_acc = history_df['val_accuracy']
plt.axis('on')
ax1 = sns.lineplot(data=loss,label='training loss',palette='husl',ax=axes[0])
_ = sns.lineplot(data=val_loss,label='validation loss',palette='viridis',ax=ax1)
ax2 = sns.lineplot(data=acc,label='training acc',palette='husl',ax=axes[1])
_ = sns.lineplot(data=val_acc,label='validation acc',palette='viridis',ax=axes[1])
plt.legend()
plt.show()
```

### We have trained and saved the model. Now we spin it back up for testing.

Now we reload the trained model as well as the saved scalers. I did not save the scalers in this notebook but it has to be done in your project. We utilize the tensorflow *tf.keras.models.load_model* to load already trained model. It takes the *filename* used to save the model as argument. I have done a lump prediction here but for your project, we use a loop to iterate through the test samples and stream results as they arrive. Again, lets talk about this further in our next meeting.

```python
model_filename = params['model-weights']
model = tf.keras.models.load_model(model_filename)
n_inputs, n_target = scaler_x.transform(X_test.reshape(-1,X_test.shape[-1])), Y_test
n_inputs = n_inputs.reshape(X_test.shape[0],X_test.shape[1],X_test.shape[2])
n_inputs = tf.cast(n_inputs, tf.float32)
n_predictions = model.predict(n_inputs,verbose=0)

# invert the output to correspond to our target range
n_predictions = scaler_y.inverse_transform(n_predictions)
n_predictions = n_predictions.flatten()
n_predictions = [round(val) for val in n_predictions]
n_target = n_target.flatten()
print(f'>>> Number of normal predictions: {len(n_predictions)}')
```

```python
plt.rcParams['lines.linewidth'] = 5
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.figure(figsize=(30,10))
plt.xlabel(r'\emph{Index}',fontsize=35)
plt.ylabel(r'\emph{Execution cycle}',fontsize=35)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
# we plot a sample here but in practice, we show everything
x_index = np.linspace(0,len(n_target[1250:1350]-1),num=len(n_target[1250:1350]))
plt.plot(x_index,n_target[1250:1350],c='g',label=r'\emph{Target}',marker='D')
plt.plot(x_index,n_predictions[1250:1350],c='r',label=r'\emph{Predictions}',marker='x')
left, right = plt.xlim()
plt.xlim(left,right)
plt.legend(loc='best',prop={'size':30})
plt.show()
```

```python
# now make predictions with delay profile
d_inputs, d_target = util.load_data('delay')
# sacle the input
delay_inputs = scaler_x.transform(d_inputs.reshape(-1,d_inputs.shape[-1]))
delay_inputs = delay_inputs.reshape(d_inputs.shape[0],d_inputs.shape[1],d_inputs.shape[2])
delay_inputs = tf.cast(delay_inputs, tf.float32)
delay_predictions = model.predict(delay_inputs,verbose=0)
delay_predictions = scaler_y.inverse_transform(delay_predictions)
delay_predictions = delay_predictions.flatten()
delay_predictions = [round(val) for val in delay_predictions]
# flatten the target
delay_target = d_target.flatten()
#print(delay_target[0:5])
#print(delay_predictions[0:5])
print(f'>>> Number of delay predictions: {len(delay_predictions)}')
```

```python
#plot the delay outcome
plt.rcParams['lines.linewidth'] = 5
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.figure(figsize=(30,10))
plt.xlabel(r'\emph{Index}',fontsize=35)
plt.ylabel(r'\emph{Execution Cycle}',fontsize=35)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
x_index = np.linspace(0,len(delay_target-1),num=len(delay_target))
plt.plot(x_index[:200],delay_target[:200],c='g',label=r'\emph{Target}',marker='D')
plt.plot(x_index[:200],delay_predictions[:200],c='r',label=r'\emph{Predictions}',marker='x')
left, right = plt.xlim()
plt.xlim(left,right)
plt.ylim(0,40)
plt.legend(loc='best',prop={'size':30})
plt.show()
```

```python
# now make predictions for random profiles
r_inputs, random_target = util.load_data('random')
random_inputs = scaler_x.transform(r_inputs.reshape(-1,r_inputs.shape[-1]))
random_inputs = random_inputs.reshape(r_inputs.shape[0],r_inputs.shape[1],r_inputs.shape[2])
random_inputs = tf.cast(random_inputs, tf.float32)
random_predictions = model.predict(random_inputs,verbose=0)
random_predictions = scaler_y.inverse_transform(random_predictions)
random_predictions = random_predictions.flatten()
random_predictions = [round(val) for val in random_predictions]
# flatten the target
random_target = random_target.flatten()
#print(random_target[0:5])
#print(random_predictions[0:5])
print(f'>>> Number of random predictions: {len(random_predictions)}')
```

```python
#plot the random prediction accuracy
plt.rcParams['lines.linewidth'] = 5
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.figure(figsize=(30,10))
plt.xlabel(r'\emph{Index}',fontsize=35)
plt.ylabel(r'\emph{Execution Cycle}',fontsize=35)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
x_index = np.linspace(0,len(random_target-1),num=len(random_target))
plt.plot(x_index[125:325],random_target[125:325],c='g',label=r'\emph{Target}',marker='D')
plt.plot(x_index[125:325],random_predictions[125:325],c='r',label=r'\emph{Predictions}',marker='x')
left, right = plt.xlim()
plt.xlim(left,right)
plt.ylim(0,50)
plt.legend(loc='best',prop={'size':30})
plt.show()
```

### Compute the errors using absolute difference for continuous data and binary difference in categorical data

```python
# compute the errors for all the profiles
window = 5
n_errors = util.regression_error(n_target,n_predictions,window)
r_errors = util.regression_error(random_target,random_predictions,window)
d_errors = util.regression_error(delay_target,delay_predictions,window)
```

```python
#plot the prediction error of the profiles
plt.rcParams['lines.linewidth'] = 5
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.figure(figsize=(30,10))
plt.xlabel(r'\emph{Index}',fontsize=35)
plt.ylabel(r'\emph{Average Prediction Error}',fontsize=35)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
#x_index = np.linspace(0,len(random_target-1),num=len(random_target))
plt.plot(n_errors[200:500],c='g',label=r'\emph{Normal Profile}')
plt.plot(r_errors[200:500],c='r',label=r'\emph{Random Profile}')
plt.plot(d_errors[200:500],c='b',label=r'\emph{Delay Profile}')
left, right = plt.xlim()
plt.xlim(left,right)
plt.ylim(0,15)
plt.legend(loc='best',prop={'size':30})
plt.grid()
plt.show()
```

We will use the validation data error to generate the mean and variance. 

```python
mu,variance = np.mean(n_errors), np.var(n_errors) # from normal profile validation data
#print(f'>>> mean: {mu}\n>>> variance: {variance}')
r_prob = util.chebyshev_probability(mu,variance,r_errors) # probability for random profile
d_prob = util.chebyshev_probability(mu,variance,d_errors)
# check the length
#print(f'>>> Len random probabilities: {len(r_prob)}\n>>> Len delay probabilities: {len(d_prob)}')
```

```python
#plot the probability
plt.rcParams['lines.linewidth'] = 5
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.figure(figsize=(30,10))
plt.xlabel(r'\emph{Index}',fontsize=35)
plt.ylabel(r'\emph{Anomaly Level Probability} $(AL_{\rho})$',fontsize=35)
plt.xticks(fontsize=30)
plt.yticks(ticks=[0.25,0.5,0.75,1.0],labels=[r'$AL_1$',r'$AL_2$',r'$AL_3$',r'$AL_4$'],fontsize=30)
#x_index = np.linspace(0,len(random_target-1),num=len(random_target))
ax = plt.subplot(111)
p1=ax.plot(r_prob[:200],c='c')
p2=ax.plot(d_prob[200:400],c='r')
left, right = plt.xlim()
plt.xlim(left,right)
axes = plt.gca()
axes.set_ylim([0,1.0])
plt.legend((p1[0], p2[0]), (r'\emph{Random} $AL_{\rho}$', r'\emph{Delay} $AL_{\rho}$'),fontsize=30,bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=2, mode="expand", borderaxespad=0.)
#plt.legend(loc='best',prop={'size':20})
plt.grid()
plt.show()
```
